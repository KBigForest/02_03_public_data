## 인구구조 그래프 함수 구현

```python
import csv 
import matplotlib.pyplot as plt
import re
import platform as pf

def parse_district_name(district):
    district_name = re.split('[()]',district)
    return district_name[0]
def print_population(poplution):
    for i in range(len(poplution)):
        print('{0:3d}세 {1:4d}명'.format(i,poplution[i]), end =' ')
        if (i + 1) % 10 ==0:
            print()
def draw_population(district_name, poplution_list):
    if (pf.system() == 'Windows'):
        plt.rcParams['font.family'] = 'D2Coding'
    plt.style.use('ggplot')
    plt.title('{0} 인구현황'.format(district_name))
    plt.xlabel('나이')
    plt.ylabel('인구수')
    plt.bar(range(101),poplution_list)
    plt.xticks(range(0,101,10))
    plt.show()

def get_population():
    city = input('인구 구조를 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요: ')
    with open('age.csv', encoding='utf-8-sig') as f:
        data = csv.reader(f)
        next(data)
        
        poplution_list = []
        full_district_name = ''
        for row in data:
            if city in row[0]:
                full_district_name = parse_district_name(row[0])
                for data in row[3:]:
                    if ',' in data:
                        data = data.replace(',', '')
                    poplution_list.append(int(data))
                break
    print_population(poplution_list)
    draw_population(full_district_name, poplution_list)

get_population()
```

```python
import csv 
import matplotlib.pyplot as plt
import re
import platform as pf

def parse_city_name(city):
    city_name = re.split('[()]', city)
    return city_name[0]
def draw_piechart(city_name, city_poploution, voting_population):
    non_voting_population = city_poploution - voting_population
    population = [non_voting_population, voting_population]
    if (pf.system() == 'Windows'):
        plt.rcParams['font.family'] = 'D2Coding'
    color = ['tomato', 'deepblue']
    plt.pue(population, labels = ['18세 미만', '투표가능인구'], autopct = '%.1f%%', colors = color)
    plt.legend()
    plt.show()
def	get_voting_population(city):

    with open('age.csv',	encoding='utf-8-sig') as f:
        data	=	csv.reader(f)
        header	=	next(data)	#	헤더 정보 건너뜀
        voting_number_list =	[]
        city_name =	''
        city_population =	0	#	도시 전체 인구수
        voting_population =	0
        for	row	in	data:
            if	city in row[0]:
                city_population =	row[1]
                if	','	in	city_population:
                #	도시 전체 인구수에서 천단위 콤마 제거
                    city_population =	city_population.replace(',',	'')
                    city_population =	int(city_population)
                    city_name =	parse_city_name(row[0])	#	(시 구 동)	이름만 분리:	지역 번호 제거
                    for	data	in	row[21:]:
                        if	','	in	data:
                            data	=	data.replace(',','')	#	천단위 콤마 제거
                            voting_num =	int(data)
                            voting_number_list.append(voting_num)	#	각 연령대별 투표 인구수를 리스트에 추가
                            voting_population +=	voting_num #	누적된 투표 가능 인구수
                    break
print('{}	전체 인구수:{:,}명,	투표 가능 인구수:	{:,}명'.format(city_name, city_population,	voting_population))
draw_piechart(city_name,	city_population,	voting_population)
```

```python
import csv 
import matplotlib.pyplot as plt
import platform as pf

def print_population(population):
    for i in range(len(population)):
        if(i <= 9):
            print('{}~{}세: {:,}명'.format(i*10, i*10+9, population[i]),end = ', ')
        else:
            print('{}세 이상: {:,}명'.format(i*10,population[i]), end = ' ')
        if (i + 1) % 5 ==0:
            print()
    print()

def draw_gender_population(male_num_list, female_num_list):
    if (pf.system() == 'Windows'):
        plt.rcParams['font.family'] = 'D2Coding'
    plt.barh(range(len(male_num_list)), male_num_list, label = '남성')
    plt.barh(range(len(female_num_list)), female_num_list, label = '남성')
    plt.rcParams['axes.unicode_minus'] = False
    plt.legend()
    plt.show()

def calculate_population():
    with open('age.csv', encoding='utf-8-sig') as f:
        data = csv.reader(f)
        male_num_list = []
        female_num_list= []

        district = input('지역 (동) 이름을 입력하세요: ')
        for row in data:
            if district in row[0]:
                for male in row[16:27]:
                    if ',' in male:
                        male = male.replace(',','')
                    male_num_list.append(-int(male))
                    
                for female in row[19:40]:
                    if ',' in female:
                        female = female.replace(',','')
                    female_num_list.append(int(female))
                break
    print('남성 총 인구수: ', sum(male_num_list))
    print_population(male_num_list)
    print('-'*40)
    print('여성 총 인구수: ', sum(female_num_list))
    draw_gender_population(male_num_list,female_num_list)


calculate_population()
```

```python
import csv
import matplotlib.pyplot as plt
import platform
import math

def draw_scatter(city, male_list, female_list, bubble_size_list):
    if (pf.system() == 'Windows'):
        plt.rcParams['font.family'] = 'D2Coding'
    plt.figure(figsize= (8,4),dpi =100)
    plt.scatter(male_list, female_list, s = bubble_size_list, c= range(11),
                alpha = 0.5, cmap = 'jet')
    plt.colorbar()
    plt.plot(range(max(male_list)),	range(max(male_list)),	'g--')
    plt.title(city + " 지역의 남녀 인구수 비교")
    plt.xlabel('남성 인구 수')
    plt.ylabel('여성 인구 수')
    plt.show()
    
def calculate_population():
    f= open('gender.csv', encoding = 'utf-8-sig')
    data = csv.reader(f)
    male_list = []
    female_list = []
    bubble_size_list = []
    city = input('찾고 싶은 지역의 이름을 입력하세요: ')
    
            
    for row in data:
        if city in row[0]:
            for i in range(16,27):
                male_num = int(row[i].replace(',',''))
                female_num = int(row[i+13].replace(',',''))
                bubble_size_list.append(math.sqrt(male_num+female_num))
                male_list.append(male_num)
                female_list.append(female_num)
            break
    f.close()
    draw_scatter(city, male_list, female_list, bubble_size_list)

calculate_population()
```
