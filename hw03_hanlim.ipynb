# 1번 문제 

```python
import csv
import matplotlib.pyplot as plt
import platform as pf
with open('gender.csv', encoding = 'utf-8-sig') as f:
    data = csv.reader(f)
    location = ['대구광역시 중구','대구광역시 동구', '대구광역시 서구', '대구광역시 남구', '대구광역시 북구', '대구광역시 수성구', '대구광역시 달서구', '대구광역시 달성군', '대구광역시']
    male_count = 0
    female_count = 0
    count = 0
    if (pf.system() == 'Windows'):
        plt.rcParams['font.family'] = 'D2Coding'
    plt.figure(figsize=(10, 8))
    for i in location:
        for row in data:
            if i in row[0]:
                male_count = int(row[14].replace(',',''))
                female_count = int(row[27].replace(',',''))
                break
        count += 1
        population = [male_count, female_count]
        colors = ['lightblue','pink']
        plt.subplot(3,4,count)
        if i == '대구광역시':
            plt.title(f'{i} 전체')
        else:
            plt.title(f'{i}')
        plt.pie(population, labels=['남성', '여성'], colors= colors, startangle=90, autopct='%.1f%%')  
    
plt.suptitle('대구광역시 구별 남녀 인구 비율', size= 20)
```

## 2-1번 문제

```python
import matplotlib.pyplot as plt
import pandas as pd
dict_citya = {'0-9': 100, '10-19': 150, '20-29': 140, '30-39': 160, '40-49':200, '50-59':240, '60-69':195, '70-79':160, '80-89':80, '90-99':20}
dict_cityb = {'0-9': 120, '10-19': 200, '20-29': 300, '30-39': 360, '40-49':400, '50-59':300, '60-69':40, '70-79':25, '80-89':20, '90-99':10}
citya_keylist = list(dict_citya.keys())
citya_valuelist = list(dict_citya.values())
cityb_valuelist = list(dict_cityb.values())
```

```python
def draw_barchart(citya_keylist, citya_valuelist, cityb_valuelist):
    n = 1
    t = 2
    d = 10
    w = 0.8
    city1_x = [t*element + w*n for element in range(d)]
    plt.bar(city1_x, citya_valuelist)
    n = 2
    t = 2
    d = 10
    w = 0.8
    city2_x = [t*element + w*n for element in range(d)]
    plt.bar(city2_x, cityb_valuelist)
    middle_x = [(a+b)/2 for (a,b) in zip(city1_x, city2_x)]
    plt.xticks(middle_x, citya_keylist)
    plt.rcParams['font.family'] == 'D2Coding'
    plt.title('The demographics of two cities')
    plt.xlabel('age')
    plt.ylabel('Population')
    plt.legend(['City A' ,'City B'])
    plt.savefig('barchart01.png', dpi=100)
draw_barchart(citya_keylist, citya_valuelist, cityb_valuelist)
```

## 2-2번 문제

```python
import matplotlib.pyplot as plt
import pandas as pd
dict_citya = {'0-9': 100, '10-19': 150, '20-29': 140, '30-39': 160, '40-49':200, '50-59':240, '60-69':195, '70-79':160, '80-89':80, '90-99':20}
dict_cityb = {'0-9': 120, '10-19': 200, '20-29': 300, '30-39': 360, '40-49':400, '50-59':300, '60-69':40, '70-79':25, '80-89':20, '90-99':10}
citya_keylist = list(dict_citya.keys())
citya_valuelist = list(dict_citya.values())
cityb_valuelist = list(dict_cityb.values())

citya_value=0
cityb_value=0
for i in range(len(citya_valuelist)):
    if i < 1:
        pass
    else:
        citya_value += citya_valuelist[i]
        cityb_value += cityb_valuelist[i]
    
citya_total = sum(citya_valuelist)
cityb_total = sum(cityb_valuelist)
title_name = 'Voting rate'
cityb_value
cas = citya_total - citya_value
cbs = cityb_total - cityb_value
```

```python
def draw_piechart(citya_total, citya_value, cityb_total, cityb_value, title_name):
    cas = citya_total - citya_value
    cbs = cityb_total - cityb_value
    p1 = [citya_value, cas]
    p2 = [cityb_value, cbs]
    plt.rcParams['font.family'] == 'D2Coding'
    plt.subplot(1,2,1)
    plt.pie(p1, labels = ['투표가능인구',''], autopct='%.1f%%',startangle=0)
    plt.title(f'City A {title_name}')
    plt.subplot(1,2,2)
    plt.pie(p2, labels = ['투표가능인구',''], autopct='%.1f%%',startangle=0)
    plt.title(f'City B {title_name}')
    plt.savefig('Voting rate.png', dpi=100)
draw_piechart(citya_total, citya_value, cityb_total, cityb_value, title_name)
```
